```python
data = yf.download(TICKER, period=PERIOD, interval=INTERVAL, auto_adjust=True)
data = data.dropna()
data["SMA_fast"] = data["Close"].rolling(FAST).mean()
data["SMA_slow"] = data["Close"].rolling(SLOW).mean()
data = data.dropna()

fig, ax = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
ax[0].plot(data.index, data["Close"], label=f"{TICKER} Close")
ax[0].plot(data.index, data["SMA_fast"], label=f"SMA {FAST}")
ax[0].plot(data.index, data["SMA_slow"], label=f"SMA {SLOW}")
ax[0].legend(); ax[0].set_title(f"{TICKER} — Price with SMA")

ax[1].bar(data.index, data["Volume"], label="Volume")
ax[1].legend(); ax[1].set_xlabel("Date"); ax[1].set_ylabel("Volume")
plt.show()
```

```python
import os, math, warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

TICKER = "AAPL" # 股票代碼
PERIOD = "2y" # 抓取 2 年資料
INTERVAL = "1d" # 日線
FAST = 10 # 快均線
SLOW = 30 # 慢均線
RISK_PER_TRADE = 0.01 # 單筆風險 1% 資金
STOP_LOSS_PCT = 0.03 # 停損 3%
INIT_CAPITAL = 100_000 # 初始資金
```

```python
df = data.copy()

# 進出場訊號
df["long_entry"] = (df["SMA_fast"].shift(1) <= df["SMA_slow"].shift(1)) & (df["SMA_fast"] > df["SMA_slow"])
df["long_exit"] = (df["SMA_fast"].shift(1) >= df["SMA_slow"].shift(1)) & (df["SMA_fast"] < df["SMA_slow"])

# 回測狀態
capital = INIT_CAPITAL
position = 0 # 持股數
entry_price = np.nan

equity_curve = []
trade_log = []

for i, row in df.iterrows():
price = row["Close"]

# 若有部位，檢查停損
if position > 0:
if (price <= entry_price * (1 - STOP_LOSS_PCT)) or row["long_exit"]:
capital += position * price
trade_log.append(("SELL", i, price))
position = 0
entry_price = np.nan

# 無部位且有買進訊號
if position == 0 and row["long_entry"]:
risk_amount = capital * RISK_PER_TRADE
stop_price = price * (1 - STOP_LOSS_PCT)
risk_per_share = max(price - stop_price, 1e-6)
shares = math.floor(risk_amount / risk_per_share)
if shares > 0:
cost = shares * price
if cost <= capital:
capital -= cost
position = shares
entry_price = price
trade_log.append(("BUY", i, price))

# 記錄權益
equity = capital + position * price
equity_curve.append((i, equity))

eq = pd.DataFrame(equity_curve, columns=["Date", "Equity"]).set_index("Date")
ret = eq["Equity"].pct_change().fillna(0)
total_return = eq["Equity"].iloc[-1] / INIT_CAPITAL - 1
max_dd = ((eq["Equity"]/eq["Equity"].cummax())-1).min()

print(f"Final Equity: ${eq['Equity'].iloc[-1]:,.0f}")
print(f"Total Return: {total_return:.2%}")
print(f"Max Drawdown: {max_dd:.2%}")
print(f"Trades: {len(trade_log)}")

fig, ax = plt.subplots(1,2, figsize=(14,4))
ax[0].plot(eq.index, eq["Equity"]); ax[0].set_title("Equity Curve")
ax[1].hist(ret, bins=50); ax[1].set_title("Daily Returns")
plt.show()

pd.DataFrame(trade_log, columns=["Side","Time","Price"]).tail(10)
```
