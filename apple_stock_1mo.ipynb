```python
import yfinance as yf
import matplotlib.pyplot as plt

# 抓取 AAPL 最近 1 個月的股價
data = yf.download("AAPL", period="1mo", interval="1d")

# 繪製收盤價走勢圖
plt.figure(figsize=(10,5))
plt.plot(data.index, data["Close"], label="AAPL Close Price")
plt.title("Apple Stock Price - Last 1 Month")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```
